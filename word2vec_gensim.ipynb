```python
import pandas as pd
import jieba
from gensim.models.word2vec import Word2Vec

# 读取数据
data = pd.read_csv('train.csv')
corpus = data['comment'].values.astype(str)

# 文本清洗 + 分词
corpus = [jieba.lcut(corpus[index]
    .replace("，", "").replace("!", "").replace("！", "")
    .replace("。", "").replace("~", "").replace("；", "")
    .replace("？", "").replace("?", "").replace("【", "")
    .replace("】", "").replace("#", "")
) for index in range(len(corpus))]
```

```python
# 1. 使用 Skip-Gram 训练 Word2Vec 模型
model = Word2Vec(corpus, sg=1, vector_size=300, window=5, min_count=3, workers=4)
print("Skip-Gram 模型训练完成")
```

```python
# 2. 输出“环境”的词向量及其形状
vec_env = model.wv["环境"]
print("词向量：\n", vec_env)
print("\n向量形状：", vec_env.shape)
```

```python
# 3. 输出与“好吃”语义最接近的3个词
print(model.wv.most_similar("好吃", topn=3))
```

```python
# 4. 计算“好吃-美味”、“好吃-蟑螂”的相似度
sim1 = model.wv.similarity("好吃", "美味")
sim2 = model.wv.similarity("好吃", "蟑螂")
print("好吃 vs 美味：", sim1)
print("好吃 vs 蟑螂：", sim2)
```

```python
# 5. 向量运算：餐厅 + 聚会 - 安静，输出1个结果
result = model.wv.most_similar(
    positive=["餐厅", "聚会"],
    negative=["安静"],
    topn=1
)
print(result)
```
